```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.models.load_model('/content/drive/MyDrive/ML_in_civil_engineering/classifier.hdf5')
```

```python
import numpy as np
from PIL import Image
# Load the image

def predictt(imgs):
  for img in imgs:
    img.show()
  # Convert the image to RGB
    rgb_img = img.convert('RGB')
    image_shape = 150
    # Resize the image to the desired shape
    resized_img = rgb_img.resize((image_shape, image_shape))

    # Convert the image to a numpy array
    img_array = np.array(resized_img)

    # Normalize the image
    img_array = img_array / 255.0
    print(img_array.shape)
    # Reshape the array to match the input shape of the model
    input_data = img_array.reshape((1, image_shape, image_shape, 3))

    # Make predictions using the model
    predictions = model.predict(input_data)
    print(predictions)
    if(predictions<0.5):
      print("Debris")
    else:
      print("Rebars")
```

```python
imgs = [
    Image.open('100061.png'),
    Image.open('200122.png')
]
predictt(imgs)
```
